```python
import sqlite3
import jinja2
from IPython.core.display import display, HTML
from flask import Flask, request, g
from flask import render_template

conn = sqlite3.connect('Wikipedia-lsi.db')
c = conn.cursor()

from ast import literal_eval

```

```python
c.row_factory = sqlite3.Row
rows = c.execute('select * from Wikipedia').fetchall()


template_string = """   {% for row in rows %}
        {{row['id']}}, <b>{{row['title']}}</b>,
        <url>{{row['url']}}</url><br>
        {% set other_titles = row['other_titles'].replace(']','').replace('[','').split(',') %}
        {% for other in other_titles%}
        <tr>{{other}}</tr><br>
        {% endfor %}
        <br>
    {% endfor %}"""
```

```python
template = jinja2.Template(template_string)
#result = template.render(rows=rows)
result = template.render(rows=rows)


display(HTML(result))
```
